```python
import pandas as pd
import csv
```

```python
zipcode_csv = "zipcodes_nc.csv"
zipcode_df = pd.read_csv(zipcode_csv)
zipcode_df.head()
```

```python
zipcode_dropped_df = zipcode_df.drop(columns=["Zipcode name","City", "State", "County Name"], axis=1)
zipcode_dropped_df.head()
```

```python
append_data = pd.DataFrame()
print('Original DF')
print(append_data)
url_df = pd.read_csv("https://www.quandl.com/api/v3/datasets/ZILLOW/Z28315_MLPSF.csv?api_key=NK4xqspYVCMidhsauWWN") 
print('Downloaded Data')
print(url_df)
append_data = pd.concat([append_data,url_df])
print('Appended Data')
print(append_data)
```

```python
append_data = pd.DataFrame()
for i in range(len(zipcode_dropped_df)):
    try:
        url = "https://www.quandl.com/api/v3/datasets/ZILLOW/Z{}_MLPSF.csv?api_key=NK4xqspYVCMidhsauWWN".format(zipcode_dropped_df['zip'][i])
        print(url)
        try:
            url_df = pd.read_csv(url)
            url_df['zip_code'] = zipcode_dropped_df['zip'][i] 
            try:
                append_data = pd.concat([append_data,url_df])
            except:
                print('Cannot append')
        except:
            print('Could not download file')
    except:
        pass
append_data
```
