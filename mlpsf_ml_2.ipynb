```python
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split

append_Final_df = pd.DataFrame()
for i in Final_df["zip_code"].unique():
    X = Final_df[Final_df['zip_code'] == i][["Date","delta_pct_pricing_per_sqft"]]
    y = Final_df[Final_df['zip_code'] == i]["pricing"].values.reshape(-1, 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y)
    #print(X.shape, y.shape)
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    preds = model.predict(X_test)
    #print(model.coef_[0][0])
    #print(model.intercept_)
    #print(score)
 #   print(preds)
    X_test['actual_price'] = Final_df['pricing']
    X_test['probability'] = score
    X_test['zip_code'] = i
    X_test['predicted_price'] = round((X_test['Date'] * model.coef_[0][0] + X_test['delta_pct_pricing_per_sqft'] * model.coef_[0][1] + model.intercept_[0]),1)
    #print(X_test)
    append_Final_df = pd.concat([append_Final_df,X_test])
append_Final_df    
```
